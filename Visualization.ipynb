```python
import torch
import matplotlib.pyplot as plt
from data_loader import create_dataloader
```

```python
# ==== Config ====
input_path = "ICASSP2025_Dataset/Inputs/Task_2_ICASSP/"
output_path = "ICASSP2025_Dataset/Outputs/Task_2_ICASSP/"
file_names = ["B1_Ant1_f1_S0"]
```

```python
# ==== Load 1 sample ====
dataloader = create_dataloader(file_names, input_path, output_path, batch_size=1, shuffle=False)
batch = next(iter(dataloader))
x, y, mask = [b.squeeze(0) for b in batch]  # remove batch dim
```

```python

# ==== Visualize input channels ====
plt.figure(figsize=(15, 4))
titles = ["Reflectance", "Transmittance", "Distance", "Sparse-PL"]
for i in range(4):
    plt.subplot(1, 4, i+1)
    plt.title(titles[i])
    plt.imshow(x[i].numpy(), cmap="jet")
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
# ==== Visualize GT and mask ====
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Ground Truth PL")
plt.imshow(y.squeeze().numpy(), cmap="jet")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Sampling Mask")
plt.imshow(mask.squeeze().numpy(), cmap="gray")
plt.axis("off")
plt.tight_layout()
plt.show()
```
